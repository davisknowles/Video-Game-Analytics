```python
# Import the Pandas library
import pandas as pd
```

```python
# Create a reference the CSV file desired
csv_path = "Resources/purchase_data.csv"

# Read the CSV into a Pandas DataFrame
purchase_df = pd.read_csv(csv_path)
```

```python
#preview df
purchase_df.head()
```

```python
#Player Count 
#Display total number of players
player_count = len(purchase_df["SN"].unique())
count_df = pd.DataFrame({"Total Players": [player_count]})
count_df
```

```python
#Purchasing Analysis (Total)
#  Run basic calculations to obtain number of unique items, average price, total number of purchases, total revenue
unique_items = len(purchase_df["Item ID"].unique())
avg_price = purchase_df["Price"].mean()
total_purchases = purchase_df["Purchase ID"].count()
total_rev = purchase_df["Price"].sum()

#  Create a summary data frame to hold the results
purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price], "Number of Purchases": [total_purchases], "Total Revenue": [total_rev]})
#  Optional: give the displayed data cleaner formatting


#  Display the summary data frame
purchasing_analysis_df
```

```python
#Purchasing Analysis (Age)
#  Bin the purchase_data data frame by age
age_grouped = purchase_df.groupby("Age Group")
total_count_age = age_grouped["SN"].nunique()

#  Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
age_purchase_count = purchase_df.groupby(["Age Group"]).count()["Purchase ID"]
age_avg_price = purchase_df.groupby(["Age Group"]).mean()["Price"]
age_purchase_total = purchase_df.groupby(["Age Group"]).sum()["Price"]
age_avg_total = age_purchase_total / total_count_age

#  Create a summary data frame to hold the results
age_analysis_df = pd.DataFrame({"Purchase Count": age_purchase_count, "Average Purchase Price": age_avg_price, "Total Purchase Value": age_purchase_total, "Avg Total Purchase per Person": age_avg_total})
age_analysis_df
#  Optional: give the displayed data cleaner formatting

#  Display the summary data frame
```

```python
#Top Spenders
#  Run basic calculations to obtain the results in the table below
sn_group_df = purchase_df.groupby(["SN"])
#Purchase Count, Average Purchase Price, Total Purchase Value
sn_purchase_count = sn_group_df.count()["Price"]
sn_avg_price = sn_group_df.mean()["Price"]  
sn_total_value = sn_group_df.sum()["Price"]
#  Create a summary data frame to hold the results
top_spender_df = pd.DataFrame({"Purchase Count": sn_purchase_count, "Average Purchase Price": sn_avg_price, "Total Purchase Value": sn_total_value})
top_spender_df.head(5)
#  Sort the total purchase value column in descending order
sorted_df = top_spender_df.sort_values("Total Purchase Value",ascending=False)
sorted_df.head(5)
#  Optional: give the displayed data cleaner formatting


#  Display a preview of the summary data frame
```

```python
#Most Popular Items
#  Retrieve the Item ID, Item Name, and Item Price columns
id_group_df = purchase_df.loc[:, ["Item ID", "Item Name", "Price"]]
#  Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
id_group_df = id_group_df.groupby(["Item ID", "Item Name"])
pop_purchase_count = id_group_df.count()["Price"]
avg_item_price = id_group_df.mean()["Price"]
pop_total_val = id_group_df.sum()["Price"]
#  Create a summary data frame to hold the results
mo_popular_df = pd.DataFrame({"Purchase Count": pop_purchase_count, "Average Item Price": avg_item_price, "Total Purchase Value": pop_total_val})

mo_popular_df

#  Sort the purchase count column in descending order
mo_pop_sorted = mo_popular_df.sort_values("Purchase Count",ascending=False)
mo_pop_sorted.head(5)
#  Optional: give the displayed data cleaner formatting


#  Display a preview of the summary data frame
```

```python
#Most Profitbale Items
#  Sort the above table by total purchase value in descending order
mo_prof_df = mo_popular_df.sort_values("Total Purchase Value",ascending=False)
mo_prof_df.head(5)
#  Optional: give the displayed data cleaner formatting


#  Display a preview of the data frame
```
